# Clean up dataset for submission

My original dataset comprises 1000 gifs. I did not realize it would result in 701Mb of gifs and 2Gb of openface data. For this reason, the submitted dataset only comprises those gifs that OpenFace successfully processed. Among the OpenFace output, only the CSVs are preserved.

```python
%ls GIFGIF_Dataset
```

* Save a copy of the original dataset (done manually)
* Walk path of extracted OpenFace data. Add every directory name (a number from 0 to 1000) to a dictionnary. 
    * Delete everything except for the CSV and the first image `*_aligned` (used to discern who was processed by OpenFace when more than two faces).
* Walk path of gifs

```python
import glob
import os
from contextlib import suppress
from PIL import Image
import math
import json
import shutil

with open("GIFGIF_Dataset/gifs.json", "r") as fp:
    gifs_dict = json.load(fp)

openface_dirs = glob.glob("./GIFGIF_Dataset/openface/sadness/*")

for dirpath in openface_dirs:
    dirname = dirpath.split("/")[-1]
    
    if dirname not in gifs_dict:
        shutil.rmtree(dirpath)
    
    else:
        with suppress(FileNotFoundError):
            os.remove(f"{dirpath}/{dirname}.avi")
            os.remove(f"{dirpath}/{dirname}.hog")
            os.remove(f"{dirpath}/{dirname}_of_details.txt")

        BMPs = glob.glob(f"{dirpath}/{dirname}_aligned/*")
        for bmp in BMPs:
            bmp_name = bmp.split("/")[-1]
            if bmp_name != "frame_det_00_000001.bmp":
                os.remove(bmp)
            else:
                name = bmp_name.split(".")[0]
                path = "/".join(bmp.split("/")[:-1]) + f"/{name}.jpg"

                foo = Image.open(bmp)
                x, y = foo.size
                x2, y2 = math.floor(x/2), math.floor(y/2)
                foo = foo.resize((x2,y2),Image.ANTIALIAS)
                foo.save(path ,quality=20)
```

```python
import subprocess

gifs = glob.glob("./GIFGIF_Dataset/gifs/*")

count = 0
for gif in gifs:
    name = gif.split("/")[-1].split(".")[0]
    if name not in gifs_dict:
        with suppress(FileNotFoundError):
            os.remove(gif)
    else:
        print(gif)
        # subprocess.run(['gifsicle', '--lossy=10', '--colors=32', '--scale=0.5', gif, '-o', f"{gif}_1.gif"]) 
        subprocess.run(['gifsicle', '--lossy=10', '--colors=128', '--scale=0.5', gif, '-o', gif]) 
```
